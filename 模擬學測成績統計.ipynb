```python
#Hw. 模擬學測成績統計
#資工三甲 409261471 游忠諺
import matplotlib.pyplot as plt
import random
import numpy

score = []  #0~60級分各出現幾次
data = []  #12萬筆資料

m1 = []   #m1筆資料彙集後，各級分出現幾次
m2 = []
m3 = []
m4 = []
m5 = []
m6 = []
m7 = []
m8 = []

for i in range(1,120001):               # 2.至少 n 筆(n 位考生)，n 值在 12 萬以上
  data.append(random.randint(0,61))     # 1.每筆總分 0~60 分 以亂數產生當中一個整數值

# 3.統計這 n 位考生的總分，0 分有幾位、1  分有幾位 ..... 60 分有幾位 
for i in range(0,61):         
  score.append(data.count(i))

#畫出 0~60 分，各分數各有幾筆的折線圖
plt.plot(score, 'go--', linewidth=2, markersize=6)
plt.show()
print()
#=====================================================================================

#第 2 ~ 9 個成果：由 10 ~100 隨機選 8 個 數字
m = []        
for i in range(1,9):
  m.append(random.randint(10, 101))    #[low, high)
print(m)
print()
#由第一筆開始，每 m 筆彙集，取平均數
#m[0]
n1 = []
if (120000%m[0]) != 0 :
  new = 120000 - 120000%m[0]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[0]  
n = new/m[0]        #12萬筆資料切分成m[0]份
k1 = m[0]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n1.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k1))

for i in range(0,61):
  m1.append(n1.count(i))

x = []              #長條圖的x軸
for i in range(0,61):
  x.append(i)


#Q2 m[0] n1
print("每" + str(m[0]) + "為一筆，總結0~60的平均數累計值，繪出各分數之長條圖")
plt.bar(x, m1)
plt.show()
print()


#Q3 m[1] n2
n2 = []
if (120000%m[1]) != 0 :
  new = 120000 - 120000%m[1]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[1]  
n = new/m[1]        #12萬筆資料切分成m[0]份
k2 = m[1]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n2.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k2))

for i in range(0,61):
  m2.append(n2.count(i))

print("每" + str(m[1]) + "為一筆，計算平均值，四捨五入後，將結果累計進該分數的累計值中，繪出 0~60 分的累計值之長條圖")
plt.bar(x, m2)
plt.show()
print()


#Q4 m[2] n3
n3 = []
if (120000%m[2]) != 0 :
  new = 120000 - 120000%m[2]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[2]  
n = new/m[2]        #12萬筆資料切分成m[0]份
k3 = m[2]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n3.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k3))

for i in range(0,61):
  m3.append(n3.count(i))

print("每" + str(m[2]) + "為一筆，計算平均值，四捨五入後，將結果累計進該分數的累計值中，繪出 0~60 分的累計值之長條圖")
plt.bar(x, m3)
plt.show()
print()


#Q5 m[3] n4
n4 = []
if (120000%m[3]) != 0 :
  new = 120000 - 120000%m[3]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[3]  
n = new/m[3]        #12萬筆資料切分成m[0]份
k4 = m[3]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n4.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k4))

for i in range(0,61):
  m4.append(n4.count(i))

print("每" + str(m[3]) + "為一筆，計算平均值，四捨五入後，將結果累計進該分數的累計值中，繪出 0~60 分的累計值之長條圖")
plt.bar(x, m4)
plt.show()
print()

#Q6 m[4] n5
n5 = []
if (120000%m[4]) != 0 :
  new = 120000 - 120000%m[4]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[4]  
n = new/m[4]        #12萬筆資料切分成m[0]份
k5 = m[4]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n5.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k5))

for i in range(0,61):
  m5.append(n5.count(i))

print("每" + str(m[4]) + "為一筆，計算平均值，四捨五入後，將結果累計進該分數的累計值中，繪出 0~60 分的累計值之長條圖")
plt.bar(x, m5)
plt.show()
print()


#Q7 m[5] n6
n6 = []
if (120000%m[5]) != 0 :
  new = 120000 - 120000%m[5]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[5]  
n = new/m[5]        #12萬筆資料切分成m[0]份
k6 = m[5]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n6.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k6))

for i in range(0,61):
  m6.append(n6.count(i))

print("每" + str(m[5]) + "為一筆，計算平均值，四捨五入後，將結果累計進該分數的累計值中，繪出 0~60 分的累計值之長條圖")
plt.bar(x, m6)
plt.show()
print()


#Q8 m[6] n7
n7 = []
if (120000%m[6]) != 0 :
  new = 120000 - 120000%m[6]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[6]  
n = new/m[6]        #12萬筆資料切分成m[0]份
k7 = m[6]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n7.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k7))

for i in range(0,61):
  m7.append(n7.count(i))

print("每" + str(m[6]) + "為一筆，計算平均值，四捨五入後，將結果累計進該分數的累計值中，繪出 0~60 分的累計值之長條圖")
plt.bar(x, m7)
plt.show()
print()


#Q9 m[7] n8
n8 = []
if (120000%m[7]) != 0 :
  new = 120000 - 120000%m[7]      #new為12萬筆資料去掉 剩餘不到m[0]而拋棄的筆數 data[0]~data[new]

length = len(data) - 120000%m[7]  
n = new/m[7]        #12萬筆資料切分成m[0]份
k8 = m[7]           #m[0] 為加總後要算平均的除數
                    #int(length/n)+1  每份的長度
for i in range(0, length, int(length/n)):   
  n8.append(numpy.rint( (sum(data[i: i + int(length/n)])) / k8))

for i in range(0,61):
  m8.append(n8.count(i))

print("每" + str(m[7]) + "為一筆，計算平均值，四捨五入後，將結果累計進該分數的累計值中，繪出 0~60 分的累計值之長條圖")
plt.bar(x, m8)
plt.show()
print()


1.從第一張折線圖可以看出0~60級分的分佈全為隨機亂數產生，無規則可循，不過能看出介於0~60級分之間各個級分的資料筆數多集中在1850筆到2000筆
2.由隨機亂數所產生的八個整數，分別為[73, 64, 14, 11, 26, 81, 38, 66]
  每一個數所代表的是12萬筆資料每m筆彙集在一起再取平均數，將結果累進到該級分數中，最後再計算0~60級分的分佈長條圖，亂數產生八個數所以有八張圖。
  其中可以看出，當每m筆資料彙集在一起時，m值越大，所產生的長條圖就越為集中;反之，當m值越小，所產生的長條圖就越為平坦。
  根據常態分度，期望值μ也就是長條圖x軸最中間的那個位置，它決定了分佈的位置，而標準差σ代表了分佈曲線的幅度。
  另外可以看出，長條圖的兩端都趨近於0
```
